```python
power_data = pd.read_csv('power.csv')
```

```python
power_data[ power_data['country'].str.contains('Est', case=True) ]['country'].unique()
```

```python
filtered = power_data[ power_data['country'].isin(['Latvia','Lithuania','Estonia']) ]
filtered = filtered[ power_data['category'].isin ([4,12,21]) ]
filtered = filtered[ (2005 < power_data['year']) & (power_data['year'] < 2010 )]
filtered = filtered[ power_data['quantity'] >= 0 ]
filtered['quantity'].sum()
```
