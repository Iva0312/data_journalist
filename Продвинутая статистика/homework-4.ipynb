Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 



1.   Разделить данные на 80% тренировочных, 20% тестовых
2.   Составить модель линейной регрессии
3.   Посчитать ошибку модели на тестовых данных
4.   Сделать точечный график данных и полученной прямой линейной регрессии



```python
MEDV = np.array(data['MEDV']).reshape(-1,1)
LSTAT = np.array(data['LSTAT']).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    LSTAT,
    MEDV,
    test_size = 0.2
)
```

```python
model.fit(X_train, y_train)
```

```python
predicted_y = model.predict(X_test)
```

```python
print(mean_squared_error(predicted_y, y_test))
print(mean_absolute_error(predicted_y, y_test))
```

```python
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.title('Линейная регрессия от одной переменной')
x = np.linspace(-50, 200, 100)
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(0,50)
plt.ylim(0,100)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

Смоделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

```python
values = []
for i in range(1000):
    value = np.random.randint(1, 7, size=100).sum()
    values.append(value)

n, bins, patches = plt.hist(values, 20, facecolor='r')

plt.xlabel('Значение')
plt.ylabel('Количества раз')
plt.title('Моделирование 1000 раз суммы по 100 подбрасываний')
plt.show()
```

Потому что в случае с моделированием 10 подбрасываний кубика с одинаковой вероятностью выпадало одно из шести значений. В случае с моделированием суммы 100 подбрасываний вероятность, что сумма выпадений кубика при минимуме 100 и максимуме 600 будет меньше 300 и больше 400, сильно ниже - значение сохраняется примерно в этом диапазоне. Чтобы сумма значений была меньше 300, кубик должен из 100 большое количество раз выпасть гранями 1, 2 и 3; больше 400 - гранями 4,  5 или 6. Центральную предельную теорему можно применить потому, что это сумма достаточно большого числа (100) независимых друг от друга случайных величин, имеющих примерно одинаковые масштабы. Как мы видим на графике, распределение и правда близко к нормальному.
