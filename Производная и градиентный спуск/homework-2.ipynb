### Задание 3: 
Посчитайте частные производные функции в точкe
(на бумажке или в LaTeX и на numpy):

$f(x,y)=2x^2y^3 + 1/x + y^2x + 7$ в точке $(1,2)$

Производная по x:
    2y^3*2x + x^-1 + y^2 = 4xy^3 - x^-2 + y^2
Подставляем значения:
    32-1+4=35

Производная по y:
    2x^2*3y^2 + 1/x + 2xy = 6x^2y^2 + 2xy
Подставляем значения: 
    24+4=28

```python
def func(x):
    return 2*x[0]**2*x[1]**3 + 1/x[0] + x[1]**2*x[0] + 7
```

```python
x = np.array([1,2])
eps = np.sqrt(np.finfo(float).eps)
approx_fprime(x, func, [eps, eps])
```

### Задание 4:

Градиентный спуск своими руками

```python
def f(x):
    return (x[0]**2) + (x[1]**2) 
```

```python
x = np.array([100, 200])
lr = 0.1
eps = np.sqrt(np.finfo(float).eps)
```

```python
for i in range(1000):
    # Calculate gradient
    grad = approx_fprime(x, f, [eps, eps])
    
    # Update x with gradient
    x = x - lr * grad
```

```python
print("Минимум достигается в: ", x)
print("Значение функции в минимуме: ", f(x))
```

Проверка с помощью встроенной функции scipy: 

```python
res = minimize(f, x, method='nelder-mead', options={'xtol': 1e-8, 'disp': True})
```
